```python
# Dependencies
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd
import numpy as np
import plotly.express as px
import json
import datetime
import os
import random
```

```python
cwd = os.getcwd()
# load the weather data from a json
with open(cwd + "/mike_analytics/LA_weather_data.json", 'r') as myfile:
    data=myfile.read()

# parse file
data_json = json.loads(data)
print(data_json[60]['dt_iso'])
```

```python
# create a data frame
clouds = []
temp = []
dt_iso = []

# loop through hours of weather data and get interesting 
for hour in data_json:
    clouds.append(hour['clouds']['all'])
    temp.append(hour['main']['temp']-273.14)
    
    # convert date and time stamp to pandas datetime
    dt_iso.append(pd.to_datetime(hour['dt_iso'][0:19], infer_datetime_format=True, utc=True))
    
# create dictionary of data
weather_dict = { 
    "Cloud Cover [%]": clouds,
    "Temperature [째C]": temp
}
```

```python
# create dataframe
weather_df = pd.DataFrame(weather_dict, index=dt_iso)
weather_df.head()
```

```python
# plot temperature
fig = px.line(weather_df, y="Temperature [째C]")
fig.show()
```

```python
# resample with daily average temperature
daily_avg_df = weather_df.resample('d').mean()
daily_avg_df.head()
```

```python
# plot temperature
fig = px.line(daily_avg_df, y="Temperature [째C]")
fig.show()
```
